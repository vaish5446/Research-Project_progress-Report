```python
## MDR Unique and cleaned foitext

import pandas as pd
import re


file_path = "2024/d_f_mdr_ae_2024.csv"  # Update path if needed
df = pd.read_csv(file_path)


df_2024 = df.dropna(subset=['FOI_TEXT']).copy()

df_2024 = df_2024.drop_duplicates(subset=['MDR_REPORT_KEY'])


def clean_text(text):
    text = str(text).lower()                              # Lowercase
    text = re.sub(r'<[^>]+>', '', text)                   # Remove HTML tags
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)            # Remove non-ASCII
    text = re.sub(r'\(.*?\)', ' ', text)                  # Remove text in parentheses
    text = re.sub(r'\d+', ' ', text)                      # Remove numbers
    text = re.sub(r'[^a-z\s]', ' ', text)                 # Remove symbols/punctuation
    text = re.sub(r'\b[a-zA-Z]\b', '', text)              # Remove single-letter words
    text = re.sub(r'\s+', ' ', text)                      # Collapse multiple spaces
    return text.strip()


df_2024['cleaned_foitext'] = df_2024['FOI_TEXT'].apply(clean_text)

output_path = "2024/d_f_mdr_ae_2024_cleaned.csv"
df_2024.to_csv(output_path, index=False)

output_path
```

```python
## MDR Unique and cleaned foitext

import pandas as pd
import re

file_path = "2022/d_f_mdr_ae_2022.csv"  
df = pd.read_csv(file_path)

df_2022 = df.dropna(subset=['FOI_TEXT']).copy()

df_2022 = df_2022.drop_duplicates(subset=['MDR_REPORT_KEY'])

def clean_text(text):
    text = str(text).lower()                              # Lowercase
    text = re.sub(r'<[^>]+>', '', text)                   # Remove HTML tags
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)            # Remove non-ASCII
    text = re.sub(r'\(.*?\)', ' ', text)                  # Remove text in parentheses
    text = re.sub(r'\d+', ' ', text)                      # Remove numbers
    text = re.sub(r'[^a-z\s]', ' ', text)                 # Remove symbols/punctuation
    text = re.sub(r'\b[a-zA-Z]\b', '', text)              # Remove single-letter words
    text = re.sub(r'\s+', ' ', text)                      # Collapse multiple spaces
    return text.strip()

df_2022['cleaned_foitext'] = df_2022['FOI_TEXT'].apply(clean_text)

output_path = "2022/d_f_mdr_ae_2022_cleaned.csv"
df_2022.to_csv(output_path, index=False)

output_path
```
